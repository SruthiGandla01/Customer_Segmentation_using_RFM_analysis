#RFM Calculation

#Recency Calculation:

```python
# Sorting the Dates to retrieve the latest date 

df.sort_values(by=['CustomerID', 'InvoiceDate'], ascending=[True, False], inplace=True)
```

```python
# Calculating the Days since the last purchase and storing it in a new Column in the dataset
import numpy as np

df['DaysSinceLastPurchase'] = df.groupby('CustomerID')['InvoiceDate'].diff().dt.days
df
```

```python
# Recency of Customer Data can now be stored in a new DataFrame with the index readjusted to get the most recent date

recency_df = df.groupby('CustomerID')['DaysSinceLastPurchase'].min().reset_index()
recency_df.head()
```

```python
frequency_df = df.groupby('CustomerID')['InvoiceNo'].nunique().reset_index()
frequency_df.columns = ['CustomerID', 'Frequency']

frequency_df.head()
```

```python
# Calculating the Monetary value by getting the product of each UnitPrice and Quantity mentioned in an order

df['TotalPrice'] = df['Quantity'] * df['UnitPrice']
monetary_df = df.groupby('CustomerID')['TotalPrice'].sum().reset_index()
monetary_df.columns = ['CustomerID', 'Monetary']

monetary_df.head()
```

#RFM Segmentation

```python
# Assigning R, F and M scores for each customer from their respective Data Frames

recency_df.sort_values(by= ['DaysSinceLastPurchase'], ascending = True, inplace = True)
recency_df['R_Score'] = pd.qcut(recency_df['DaysSinceLastPurchase'], q=[0, 0.25, 0.5, 0.75, 1], labels=False,\
                                duplicates = 'drop') + 1
recency_df['R_Score'].fillna(0.0, inplace = True)

recency_df
```

```python
frequency_df.sort_values(by= ['Frequency'], ascending = False, inplace = True)
frequency_df['F_Score'] = pd.qcut(frequency_df['Frequency'], q=[0, 0.25, 0.5, 0.75, 1], labels=False, duplicates='drop') + 1

frequency_df
```

```python
monetary_df.sort_values(by= ['Monetary'], ascending = False, inplace = True)
monetary_df['M_Score'] = pd.qcut(monetary_df['Monetary'], q=[0, 0.25, 0.5, 0.75, 1], labels=False, duplicates='drop') + 1

monetary_df
```

```python
# Combining the R, F and M scores to create a single score
# The formula used to combine these scores : R * 100 + F * 10 + M

rfm_scores = pd.merge(recency_df[['CustomerID', 'R_Score']], frequency_df[['CustomerID', 'F_Score']], on='CustomerID')
rfm_scores = pd.merge(rfm_scores, monetary_df[['CustomerID', 'M_Score']], on='CustomerID')

rfm_scores['RFM_Score'] = rfm_scores['R_Score'] * 100 + rfm_scores['F_Score'] * 10 + rfm_scores['M_Score']
rfm_scores
```

#Customer Segmentation

```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Assuming you have a DataFrame named 'rfm_scores' with 'RFM_Score', 'R_Score', 'F_Score', 'M_Score'

# Select the relevant columns for clustering
X = rfm_scores[['RFM_Score', 'R_Score', 'F_Score', 'M_Score']]

# Standardize the variables
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Specify the number of clusters 
num_clusters = 4

# Apply K-Means clustering
kmeans = KMeans(n_clusters=num_clusters, random_state=42,n_init=10)
rfm_scores['Cluster'] = kmeans.fit_predict(X_scaled) + 1  # Adding 1 to start cluster numbering from 1

# Display the updated DataFrame with cluster assignments
print(rfm_scores.head())

```

```python
# Experiment with different numbers of clusters
import matplotlib.pyplot as plt
cluster_range = range(1, 11)
inertia = []

for num_clusters in cluster_range:
    kmeans = KMeans(n_clusters=num_clusters, random_state=42,n_init=10)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)

# Plot the Elbow curve
plt.figure(figsize=(8, 6))
plt.plot(cluster_range, inertia, marker='o')
plt.title('Elbow Method for Optimal Number of Clusters')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia (Within-cluster sum of squares)')
plt.show()
```

#Segment Profiling

```python
# Assuming you have a DataFrame named 'rfm_scores' with 'CustomerID', 'RFM_Score', 'R_Score', 'F_Score', 'M_Score', and 'Cluster'

# Extract cluster information
cluster_info = rfm_scores[['CustomerID', 'RFM_Score', 'R_Score', 'F_Score', 'M_Score', 'Cluster']]
```

```python
# Group by cluster and calculate summary statistics
cluster_summary = cluster_info.groupby('Cluster').agg({
    'CustomerID': 'count',
    'RFM_Score': ['mean', 'min', 'max'],
    'R_Score': ['mean', 'min', 'max'],
    'F_Score': ['mean', 'min', 'max'],
    'M_Score': ['mean', 'min', 'max']
}).reset_index()

# Display the cluster summary
print(cluster_summary)
```

```python
# Visualize RFM scores for each cluster
import seaborn as sns
for col in ['RFM_Score', 'R_Score', 'F_Score', 'M_Score']:
    plt.figure(figsize=(10, 6))
    sns.boxplot(x='Cluster', y=col, data=cluster_info)
    plt.title(f'{col} Distribution for Each Cluster')
    plt.show()
```
